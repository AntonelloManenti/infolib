```python
#pd.Dataframe test 1

test_01 = {'A': ['Lorem ipsum', 'Sed', 1], 'B': [0.2, 4.5442233, 1.3], 
  'C': [0, 1, 12345], 
  'D': [False, True, False], 'E': [2000, 2001, 2003],
  'F': [np.nan, np.nan, np.nan], 'G': ['A', 'B', 'C'], 'H':[np.nan, np.nan, np.nan]}

test_01 = pd.DataFrame.from_dict(test_01)
```

```python
#pd.Dataframe test 2 - no pd.Dataframe

test_02 = ['a', 'b', 1, 3]
```

```python
#pd.Dataframe test 3 - all dtype

s = pd.Series(pd.date_range("2012-1-1", periods=3, freq="D"))
td = pd.Series([pd.Timedelta(days=i) for i in range(3)])
i = [1,2,3]
f = [0.123,423.231,0.002]
c = ['A', 'B', 'C']
cn = [1, 2, 3]
b = [False, True, False]
n = [np.nan, np.nan, np.nan]

test_03 = pd.DataFrame({"A": s, "B": td, "C": i, "D":f, "E":c, "F":cn, "G":b, "H":n})
test_03['E'] = test_03['E'].astype('category')
test_03['F'] = test_03['F'].astype('category')
```

```python
%%time

info(test_01)
```

```python
%%time

info(test_02)
```

```python
%%time

info(test_03)
```
